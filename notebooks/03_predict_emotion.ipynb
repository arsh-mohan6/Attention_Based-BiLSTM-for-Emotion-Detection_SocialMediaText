```python
import numpy as np
import pickle
from tensorflow.keras.models import load_model
from tensorflow.keras.preprocessing.sequence import pad_sequences
```

```python
model = load_model("/content/drive/MyDrive/emotion_project/models/attention_bilstm_model.keras")

print("Model loaded successfully")
```

```python
import pickle

with open("/content/drive/MyDrive/emotion_project/models/tokenizer.pkl", "rb") as f:
    tokenizer = pickle.load(f)

print("Tokenizer loaded successfully")
```

```python
emotion_map = {
    0: "sadness",
    1: "joy",
    2: "love",
    3: "anger",
    4: "fear",
    5: "surprise"
}
```

```python
text = input("Enter a sentence: ")

# convert text to sequence
seq = tokenizer.texts_to_sequences([text])

# pad sequence
padded = pad_sequences(seq, maxlen=50)

# predict emotion
pred = model.predict(padded)

label = np.argmax(pred)
pred = model.predict(padded)

for i, prob in enumerate(pred[0]):
    print(emotion_map[i], ":", round(prob, 3))

label = np.argmax(pred)
print("\nPredicted Emotion:", emotion_map[label])
```
